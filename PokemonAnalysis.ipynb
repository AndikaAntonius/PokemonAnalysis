```python
#importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#importing the data
df = pd.read_csv("/content/Pokemon.csv")
df
```

```python
#Get information about DataFrame structure
df.info()
```

```python
# Generate descriptive statistics of DataFrame.
df.describe()
```

```python
type1 = pd.unique(df['Type 1'])
type1
```

```python
type2 = pd.unique(df['Type 2'])
type2
```

```python
#Counting the number of Pokémon for each 'Type 2'
pokemon_per_type = df['Type 1'].value_counts()

#Sorting the counts of Pokémon per 'Type 2'
pokemon_per_type_sorted = pokemon_per_type.sort_values(ascending=True)

# Creating a horizontal bar plot
plt.figure(figsize=(8, 5))
bars = pokemon_per_type_sorted.plot(kind='barh', color='skyblue', width=0.8)

# Adding count labels on each bar
for bar in bars.patches:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2,
             f'{int(bar.get_width())}', ha='left', va='center')
# Creating a bar plot
plt.title('Pokémon Type 1')
plt.xlabel('Count')
plt.ylabel('Type')
plt.tight_layout()
plt.show()

```

```python
#Counting the number of Pokémon for each 'Type 2'
pokemon_per_type = df['Type 2'].value_counts()

#Sorting the counts of Pokémon per 'Type 2'
pokemon_per_type_sorted = pokemon_per_type.sort_values(ascending=True)

# Creating a horizontal bar plot
plt.figure(figsize=(8, 5))
bars = pokemon_per_type_sorted.plot(kind='barh', color='yellow', width=0.8)

# Adding count labels on each bar
for bar in bars.patches:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2,
             f'{int(bar.get_width())}', ha='left', va='center')

# Creating a bar plot
plt.title('Pokémon Type 2')
plt.xlabel('Count')
plt.ylabel('Type')
plt.tight_layout()
plt.show()
```

```python
#Dropping non-numeric columns
df_numeric = df.drop(columns=['#', 'Name', 'Type 1', 'Type 2', 'Legendary'])

# Creating a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_numeric.corr(), cmap='coolwarm', annot=True)
plt.title('Heatmap of Correlation')
plt.tight_layout()
plt.show()

```

```python
# Sorting the DataFrame df by the 'HP' column
df_sorted = df.sort_values(by=['HP'], ascending=False)

# Selecting the top 5 rows
top_5 = df_sorted.head()

# Creating a bar plot
plt.figure(figsize=(10, 6))
plt.bar(top_5['Name'], top_5['HP'], color='red')
plt.xlabel('Pokémon', rotation=0)
plt.ylabel('HP')
plt.title('Top 5 HP of Pokémon')
plt.xticks(rotation=0)

# Adding count labels on each bar
for p in plt.gca().patches:
    height = p.get_height()
    plt.gca().text(p.get_x() + p.get_width() / 2, height + 3,
                   f'{int(height)}', ha='center', color='black', fontsize=10)
plt.show()

```

```python
# Selecting specific columns
selected_columns = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
selected_df = df[selected_columns]

# Creating a boxplot for the selected attributes
plt.figure(figsize=(8, 8))
sns.boxplot(data=selected_df)
plt.title('Distribution of Pokémon Attributes')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Boxplot untuk atribut 'Total'
plt.figure(figsize=(2, 6))
sns.boxplot(y=df['Total'], color='skyblue')
plt.title('Distribution of Pokémon Total')
plt.xlabel('Total')
plt.tight_layout()
plt.show()

```

```python
# Calculate the number of Pokémon per generation
pokemon_per_generation = df['Generation'].value_counts().sort_index()

# Sort the number of Pokémon per generation from highest to lowest
pokemon_per_generation_sorted = pokemon_per_generation.sort_values(ascending=False)

# Creating a plot
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Generation',
              order=pokemon_per_generation_sorted.index, palette='viridis', hue='Generation')
plt.title('Pokémon per Generation')
plt.xlabel('Generation')
plt.ylabel('Count')

# Adding count labels on each bar
for p in plt.gca().patches:
    height = p.get_height()
    plt.gca().text(p.get_x() + p.get_width() / 2, height + 3,
                   f'{int(height)}', ha='center', color='black', fontsize=10)

# Displaying legend
plt.legend(title='Generation', loc='upper right')

plt.tight_layout()
plt.show()
```

```python
# Calculating the total number of Pokémon
total_pokemon = len(df)

# Calculating the number of Legendary Pokémon
num_legendary = df['Legendary'].value_counts()[True]
num_non_legendary = total_pokemon - num_legendary

# Data for the bar chart
labels = ['TRUE', 'FALSE']
sizes = [num_legendary, num_non_legendary]
colors = ['#FFD700', '#ADD8E6']

# Creating the bar chart
plt.figure(figsize=(6, 6))
plt.bar(labels, sizes, color=colors)
plt.title('Number of Legendary Pokémon')
plt.xlabel('Legendary')
plt.ylabel('Count')

# Adding count labels on each bar
for p in plt.gca().patches:
    height = p.get_height()
    plt.gca().text(p.get_x() + p.get_width() / 2, height + 3,
                   f'{int(height)}', ha='center', color='black', fontsize=10)

plt.show()



```
